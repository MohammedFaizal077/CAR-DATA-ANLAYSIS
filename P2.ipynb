THE PREDICTION USING MILEAGE WHEN CC UIS GIVEN 

```python
import numpy as np

x = df["engine"]
y = df["mileage"]

m = np.sum((x - x.mean()) * (y - y.mean())) / np.sum((x - x.mean())**2)
c = y.mean() - m * x.mean()

# prediction
engine_value = float(input("Enter engine (cc): "))
predicted_mileage = m * engine_value + c

print("THE PREDICTED MILEAGE FOR AN ENGINE OF", engine_value, "CC IS:", predicted_mileage)
```

## 1.How does car age affect selling price?

- Answerd this question by showing the scatter plot and line graph(using groupby function) of the cateogory of price and year of the car 

The code and the graph and findings are given below 

```python
year_price=df.groupby("year")["selling_price"].mean()
plt.plot(year_price.index,year_price.values,marker= 'o')
plt.title('year vs selling price')
plt.xlabel("year")
plt.ylabel("average price")
plt.show()
```

```python


plt.figure(figsize=(8,5))
plt.scatter(df["year"], df["selling_price"], alpha=0.5)
plt.xlabel("Manufacturing Year")
plt.ylabel("Selling Price")
plt.title("Relationship between Car Age and Selling Price")
plt.grid(True)
plt.show()
```

## FINDINGS

- Negative relationship
→ As car age increases, selling price decreases.
- Newer cars have higher prices
→ Recently manufactured cars are more expensive.
- Older cars have lower prices
→ Older vehicles are sold at cheaper rates.
- Depreciation effect
→ Car value reduces over time due to usage and wear.
- Higher price variation in newer cars
→ Prices of newer cars vary widely based on features and condition.
- Lower price concentration in older cars
→ Older cars mostly fall within a lower price range
