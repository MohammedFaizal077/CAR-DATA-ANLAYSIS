REMOVING THE UNITS FROM ENGINE ,MILEAGE FOR NUMERICAL OPERATIONS

```python
df[["engine","engine_unit"]]=df["engine"].str.split(" ",expand=True)
```

```python
df
```

```python
df[["mileage","mileage_unit"]]=df["mileage"].str.split(" ",expand=True)
```

```python
df
```

```python
df[["max_power","max_power_unit"]]=df["max_power"].str.split(" ",expand=True)
```

```python
df
```

NOW MAKE ALL THE NUMERICASL VALUES TO INTEGER\FLOAT

```python
df['mileage'] = df['mileage'].astype(float)
```

```python
df['engine'] = df['engine'].astype(int)
```

```python
df['max_power'] = df['max_power'].astype(float)
```

```python
df
```

ALL INFORMATION ABOUT THE DATA SET AFTER CLEANING

```python
df.describe()
```

THERE HAS MILEAGE 0 SO NEED TO CONVERT IT 

```python
df['mileage'] = df['mileage'].replace(0, df['mileage'].mean())
```
